```python
N = 1000
BF_hd_curn = np.zeros(N)
BF= np.zeros(N)
for i in range(N):

    hd_z = np.random.choice(hd)
    curn_z = np.random.choice(curn)
    BF_hd_curn[i] = hd_z - curn_z

for i in range(N):

    irn_z = np.random.choice(irn)
    spl_z = np.random.choice(hd)
    BF[i] = spl_z - irn_z 

ymax = 0.3

plt.hist(BF_hd_curn,bins= 20,alpha=1,density=True,histtype='step',color='green',label='GSS')
plt.vlines(np.mean(BF_hd_curn), ymin=0, ymax=ymax,colors='green',alpha=0.7),#label='mean BF_Curn_IRN') 
plt.vlines(np.mean(BF_hd_curn)-np.std(BF_hd_curn), ymin=0, ymax=ymax,colors='green',linestyle='dashed',alpha=0.6)
plt.vlines(np.mean(BF_hd_curn)+np.std(BF_hd_curn), ymin=0, ymax=ymax,colors='green',linestyle='dashed',alpha=0.5)

nano=np.log(226)
nano_err = np.log(70)
#plt.vlines(np.median(BF), ymin=0, ymax=0.1,colors='black',label='median BF_HD_Curn') 
plt.vlines(nano, ymin=0, ymax=ymax,colors='grey',label='Nanograv')

plt.title('log(BF) distribution of GWB vs CURN model for Nanograv')
plt.vlines(nano-nano_err, ymin=0, ymax=ymax,colors='black',linestyle='dashed',alpha=0.7)
plt.vlines(nano+nano_err, ymin=0, ymax=ymax,colors='black',linestyle='dashed',alpha=0.7)
plt.legend()
plt.xlabel('log(BF)')
plt.ylabel('PDF')
print(f'{N} log_BF_rn realisation mean:{np.mean(BF)} , std:{np.std(BF)}')
```
