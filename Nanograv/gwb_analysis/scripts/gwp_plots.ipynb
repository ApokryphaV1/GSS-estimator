```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 130

# Combine all vectors into a list
irn = np.genfromtxt('../paper_results/irn_tf100_5000s_log_z_100_16T.txt',dtype=float,unpack=True) 
curn =  np.genfromtxt('../paper_results/curn_tf100_5000s_log_z_100_16T.txt',dtype=float,unpack=True)
#HD 35gb 8T 3h:3min
spl = np.genfromtxt('../paper_results/spl_tf100_5000s_log_z_100_16T.txt',dtype=float,unpack=True)
hd = np.genfromtxt('../paper_results/hd_tf100_5000s_log_z_100_16T.txt',dtype=float,unpack=True)

data = [ curn , spl , hd]
#data = [irn, curn , spl , hd]
box_color = 'black'

plt.boxplot(data,boxprops=dict(color=box_color, alpha=1),     # Change the box color
            whiskerprops=dict(color=box_color, alpha=1), # Change whisker color
            capprops=dict(color=box_color, alpha=1),   # Change cap color
            medianprops=dict(color=box_color),showfliers=False)

#plt.hlines(np.log(3),xmin=0.5,xmax=4.5, color='red', linestyle='--', label='Threshold: Substantial Evidence')
#plt.hlines(np.log(10),xmin=0.5,xmax=4.5, color='orange', linestyle='--', label='Threshold: Strong Evidence')

# Set x-labels
#plt.plot([],[],box_color,label='CURN')

#plt.plot([],[],hd_color, alpha=0.5,label='HD')

plt.xticks([ 1, 2 , 3], [ 'CURN','SPL','HD'])
#plt.hlines(np.median(hd2_32t),xmin=0,xmax=3,linestyles='dashed',colors='red')
# Set labels and title
plt.xlabel('HD with 1 samples per temperatures')
plt.ylabel('Log($Z$)')
#plt.title('HD Nanograv Log(Z)')
plt.legend()
# Display the plot
plt.show()
```

```python
N = 1000
BF_hd_curn = np.zeros(N)
BF_hd_curn = np.zeros(N)

BF= np.zeros(N)
for i in range(N):

    hd_z = np.random.choice(hd)
    curn_z = np.random.choice(curn)
    BF_hd_curn[i] = hd_z - curn_z

for i in range(N):

    m2_z = np.random.choice(curn)
    m1_z = np.random.choice(hd)
    BF[i] = m1_z - m2_z 

ymax = 0.3

plt.hist(BF_hd_curn,bins= 20,alpha=1,density=True,histtype='step',color='green',label='GSS')
plt.vlines(np.mean(BF_hd_curn), ymin=0, ymax=ymax,colors='green',alpha=0.7),#label='mean BF_Curn_IRN') 
plt.vlines(np.mean(BF_hd_curn)-np.std(BF_hd_curn), ymin=0, ymax=ymax,colors='green',linestyle='dashed',alpha=0.6)
plt.vlines(np.mean(BF_hd_curn)+np.std(BF_hd_curn), ymin=0, ymax=ymax,colors='green',linestyle='dashed',alpha=0.5)

nano=np.log(226)
nano_err = np.log(70)
#plt.vlines(np.median(BF), ymin=0, ymax=0.1,colors='black',label='median BF_HD_Curn') 
plt.vlines(nano, ymin=0, ymax=ymax,colors='grey',label='Nanograv')

plt.title('log(BF) distribution of GWB vs CURN model for Nanograv')
plt.vlines(nano-nano_err, ymin=0, ymax=ymax,colors='black',linestyle='dashed',alpha=0.7)
plt.vlines(nano+nano_err, ymin=0, ymax=ymax,colors='black',linestyle='dashed',alpha=0.7)
plt.legend()
plt.xlabel('log(BF)')
plt.ylabel('PDF')
print(f'{N} log_BF_rn realisation mean:{np.mean(BF)} , std:{np.std(BF)}')
```
